```python
import numpy as np
from scipy.linalg import hilbert
import matplotlib.pyplot as plt
```

### Function to compute the error

```python
def error(A,x_true):
    b = A @ x_true
    x=np.linalg.solve(A,b)
    error=np.linalg.norm(x-x_true,2)/np.linalg.norm(x_true,2)
    return error
```

### Plot of errors and Condition number of random matrices with different size "n"

```python
n_fun=np.arange(10,110,10)
errors=[error(np.random.rand(i,i),np.ones(i)) for i in n_fun ]
plt.plot(n_fun,errors,color="violet")
plt.xlabel("n")
plt.ylabel("error")
plt.title("graph with relative errors as a function of n with Random matrix",color="blue")
plt.grid()
plt.show()
conds_2=[np.linalg.cond(np.random.rand(i,i),2) for i in n_fun ]
conds_inf=[np.linalg.cond(np.random.rand(i,i),np.inf) for i in n_fun ]
plt.plot(n_fun,conds_2)
plt.plot(n_fun,conds_inf)
plt.legend(["Cond-Norm-2","Cond-Norm-inf"])
plt.show()
```

#### Plot of errors and Condition number of Vandermonde matrices with different size "n"

```python
n_fun_vander=np.arange(5,35,5)
errors_vander=[error(np.vander(np.array(range(1,i+1))),np.ones(i)) for i in n_fun_vander]
plt.plot(n_fun_vander,errors_vander,color="violet")
plt.xlabel("n")
plt.ylabel("error")
plt.title("graph with relative errors as a function of n with Vander matrix",color="blue")
plt.grid()
plt.show()
conds_2_vander=[np.linalg.cond(np.vander(np.array(range(1,i+1))),2) for i in n_fun_vander ]
conds_inf_vander=[np.linalg.cond(np.vander(np.array(range(1,i+1))),np.inf) for i in n_fun_vander ]
plt.plot(n_fun_vander,conds_2_vander)
plt.plot(n_fun_vander,conds_inf_vander)
plt.legend(["Cond-Norm-2","Cond-Norm-inf"])
plt.show()
```

#### Plot of errors and Condition number of Hilbert matrices with different size "n"

```python
n_fun_hilbert=np.arange(4,13,1)
errors_hilbert=[error(hilbert(i),np.ones(i)) for i in n_fun_hilbert]
plt.plot(n_fun_hilbert,errors_hilbert,color="violet")
plt.xlabel("n")
plt.ylabel("error")
plt.title("graph with relative errors as a function of n with Hilbert matrix",color="blue")
plt.grid()
plt.show()
conds_2_hilbert=[np.linalg.cond(np.vander(np.array(range(1,i+1))),2) for i in n_fun_hilbert ]
conds_inf_hilbert=[np.linalg.cond(np.vander(np.array(range(1,i+1))),np.inf) for i in n_fun_hilbert ]
plt.plot(n_fun_hilbert,conds_2_hilbert)
plt.plot(n_fun_hilbert,conds_inf_hilbert)
plt.legend(["Cond-Norm-2","Cond-Norm-inf"])
plt.show()
```
